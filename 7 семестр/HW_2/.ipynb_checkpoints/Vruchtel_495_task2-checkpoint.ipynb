### 3. Тест на стационарность

```python
kpss_stat= stattools.kpss(np.array(ts_boxcox))
print('The KPSS test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ntable distribution: %s' % 
      (kpss_stat[0], kpss_stat[1], kpss_stat[2], kpss_stat[3]))
```

```python
df_stat = stattools.adfuller(np.array(ts_boxcox))
print('The Augmented Dickey-Fuller test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ncritical values: %s' % 
      (df_stat[0], df_stat[1], df_stat[2], df_stat[4]))
```

По результатам тестов видно, что ряд нельзя считать стационарным. Значит, его нужно дифференцировать.

```python
# Устраним недельную сезонность
p=7
dif7_ts = (ts_boxcox - ts_boxcox.shift(p))[p:]
ts_boxcox.plot(figsize=(15,5), label="without dif7")
dif7_ts.plot(figsize=(15,5), label="with dif7")
plt.legend()
```

```python
# Проверим новый полученный ряд на стационарность
kpss_stat= stattools.kpss(np.array(dif7_ts))
print('The KPSS test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ntable distribution: %s' % 
      (kpss_stat[0], kpss_stat[1], kpss_stat[2], kpss_stat[3]))
```

### 4. Изобразим ACF и PACF.

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(dif50_ts.values.squeeze(), lags=50, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(dif50_ts.values.squeeze(), lags=50, ax=ax2)
```

### 6. Обучим мдели ARIMA с выбранными параметрами, выберем наилучшую.

```python
def train_arima_model(p, d, q):
    ## Choosing best model accoridng to AIC criterion
    arima_model = sm.tsa.ARIMA(dif50_ts, (p,d,q)).fit()
    #print(arima_model.params)
    print("ARIMA (", p, ",", d, ",", q, ")")
    print("AIC: ", arima_model.aic)
    return arima_model
```

### 3. Grid search for ARIMA parameters

```python
def find_best_arima():
    
    AIC_to_params = {}
    
    # grid search
    for p in [3, 5, 7, 10, 12]:
        for q in [3, 5, 7, 10, 12]:
            for d in [0, 1, 2, 3]:
                try:
                    arima_mod_cur = train_arima_model(p, d, q)
                    AIC_to_params[arima_mod_cur.aic] = [p, d, q]
                except Exception:
                    print("p:", p, "d:", d, "q:", q, "ERROR!")
                    
    print("")
    # find best arima
    min_AIC = min(AIC_to_params)
    
    best_params = AIC_to_params[min_AIC]
    print("BEST PARAMS:", best_params)
    print("AIC:", min_AIC)
    
    arima_mod_best = train_arima_model(best_params[0], best_params[1], best_params[2])
    
    initial_ts = my_ts
    # Применим преобразование boxcox
    initial_ts_boxcox = np.log(initial_ts + 1)
    # Устраняем сезонность
    p = 50
    initial_dif50_ts = (initial_ts_boxcox - initial_ts_boxcox.shift(p))[p:]
    
    len_to_forecast = len(my_ts) - len(my_ts_train)
    
    forecast = arima_mod_best.forecast(steps=len_to_forecast)
    
    plt.figure(figsize=(15, 5))
    plt.plot(range(len_to_forecast), forecast[0], label="forecast")
    plt.fill_between(range(len_to_forecast), forecast[0] - forecast[1], forecast[0] + forecast[1], alpha=0.2)
    plt.plot(range(-len(initial_dif50_ts) + len_to_forecast, len_to_forecast),
             initial_dif50_ts, label="initial")
    plt.legend()
    plt.show()
    
    print('MACAPE: %.2f'% qualityMACAPE(initial_dif50_ts[-len_to_forecast:], forecast[0])[0])
    
```

```python
find_best_arima()
```
