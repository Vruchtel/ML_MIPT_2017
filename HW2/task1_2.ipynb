### Вручтель Серафима. 495. Линейные модели в деревьях.

Условие:

Одна из частых идей попытаться улучшить регрессионное дерево, выдавая вместо константных ответов в листьях ответ линейной регрессии, обученной на объектах из этого листа. Как правило такая стратегия не дает никакого ощутимого выигрыша. Попробуйте объяснить, почему? Как стоит модифицировать построение разбиений в дереве по MSE, чтобы при разбиении получались множества, на которых линейные модели должны работать неплохо?


Решение:
    
Возможно, стратегия использования в листьях дерева линейной регрессии не даёт никакого ощутимого выигрыша, поскольку, если считать, что в каждом из сплитов мы выбираем наилучшее разбиение, то с высокой вероятностью может оказаться, что окажется даже лучшее разбиение, чем линейная регрессия.

Могу предложить следующую модификацию построения разбиений в дереве: можно в каждом сплите проверять, насколько хорошо здесь работает линейная модель, и, если она работает достаточно хорошо (согласно каким-нибудь критериям оценки), то дальше не разбивать. То есть сделать данный сплит в ветке последним.
