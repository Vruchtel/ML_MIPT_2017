**Вопрос 1**: В чём заключается проблема мультиколлинеарности?

** Вопрос 4**: Какая связь между решением задачи PCA и SVD-разложением матрицы регрессии?

Ответ: В PCA (методе главных компонент) матрица переменных $X$ разлагается в произведение матриц $T$ и $P$. В SVD (разложение по сингулярным значениям) матрица переменных $X$ разлагается в произведение трех матриц: $X = USV^t$, где $T=US$, $P = V$.

** Вопрос 5**: Почему в tSNE расстояние между парами объектов измеряется "по-стьюденту" и как это помогает решить проблему "скрученности" (crowding problem)?

Ответ: Проблема скрученности заключается в том, что расстояние между двумя точками в изображении, соответствующими двум среднеудалённым точкам в исходном многомерном пространстве, должно быть существенно больше, чем расстояние, которое позволяет получить гауссово распределение (используется в SNE). Эту проблему решают тяжёлые хвосты распределения Стьюдента.

**Вопрос 6**: На какой идее базируются алгоритмы агломеративной кластеризации? Напишите формулу Ланса-Вильма

Ответ: алгоритмы агломеративной кластеризации базируются на идее объединения более мелких кластеров в более крупные до тех пор, пока эти кластеры не станут достаточно "хорошими". Формула Ланса-Вильма имеет следущий вид:

$R(U \cup V, S) = \alpha_U \cdot R(U, S) + \alpha_V \cdot R(V, S) + \beta \cdot R(U, V) + \gamma \cdot |R(U, S) - R(V, S)|$, где $\alpha_U$, $\alpha_V$, $\beta$, $\gamma$ - числовые параметры, $R(U \cup V, S)$ - расстояние между кластерами $W = U \cup V$ и $S$.

**Вопрос 8**: В чём отличия (основные упрощения) k-means от EM-алгоритма кластеризации?
 
Ответ: EM-алгоритм основан на предположении, что данные подчиняются нормальному распределению, и элемент относится к кластеру, вероятность попадания в который выше, в k-means же элемент распределяется в кластер, расстояние до которого минимально. Кроме того, в EM форма кластеров эллиптическая и настраиваемая, а в k-means форма кластеров жёстко зависит от метрики.

**Вопрос 9** Какой принцип работы графовых алгоритмов кластеризации?

Ответ: суть графовых алгоритмов заключается в том, что выборка представляется в виде графа, вершинами которого служат объекты, а рёбра - расстояния между объектами. 

**Вопрос 10**  В чем некорректность постановки задачи кластеризации?

Ответ: некорректность постановки задачи кластеризации заключается в том, что во-первых существует великое множество способов оценки качества кластеризации, и не понятно, какой из них считать правильным, во-вторых, как правило, неизвестно число кластеров, на которые требуется разбить данные, и неизвестно, какое число считать наилучшим, кроме того, результат кластеризации сильно зависит от метрики расстояния между кластерами, которая задаётся субъективно.

```python
import numpy as np
import pylab as plt
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    Sigma = 1.0 / l * np.dot(F, np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(Sigma)

    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / dTot * 100.0 for i in range(n)])
    # вычислите необходимое число главных компонент
    m = np.argmax(var_i > 100.0 * varRetained)
    # верните число m и точность, которая достигается при этом числе главных компонент
    print '%.2f %% variance retained in %d dimensions' % (var_i[m], m)

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    # только m главных компонент
    U_reduced = U[:m]
    # вычислить матрицу в преобразованном пространстве
    G = np.dot(U_reduced, F)

    return G, U_reduced
```

```python
# Кластеризация цифр с помощью DBSCAN

from sklearn.datasets import load_digits
from sklearn.cluster import DBSCAN

from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
# Загрузим данные

X, y = load_digits(return_X_y=True)
X_train, X_test, Y_train, Y_test = train_test_split(X, y, train_size=0.85)
```

```python
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
clusters_pca = DBSCAN().fit(X_train_pca)
print clusters_pca.labels_
```
